# Cardiovascular disease data
This data comes from the World Health Organization (http://www.who.int/gho/database/en/).  Cardiovascular disease is an example of a non-communicable disease, meaning it is not contracted from other people, although genetic factors may be involved.  Lifestyle choices such as diet and exercise are thought to be important factors in whether someone develops cardiovascular disease.

```python
#Load cardiovascular disease data
cardio_file = '../data/data.csv'
cardio_data = pd.read_csv(cardio_file)

#Select relevant field
cardio_data = cardio_data.drop(['Age-standardized mortality rate by cause (per 100 000 population).1', \
                  'Age-standardized mortality rate by cause (per 100 000 population).2'], axis=1)

#Separate 2000 data.  These will be used for model training
cardio_2000 = cardio_data[cardio_data['Unnamed: 1'].isin(['2000'])]
cardio_2000 = cardio_2000.drop(['Unnamed: 1'], axis=1)
#These are countries
cardio_2000 = cardio_2000.rename(columns={'Unnamed: 0' : 'Country'})

#To print all country names
# for item in cardio_2000.index:
#     print item

#Repeat to get 2012 data, which will be used for model testing
cardio_2012 = cardio_data[cardio_data['Unnamed: 1'].isin(['2012'])]
cardio_2012 = cardio_2012.drop(['Unnamed: 1'], axis=1)
#These are countries
cardio_2012 = cardio_2012.rename(columns={'Unnamed: 0' : 'Country'})

#Show dataframe head
cardio_2012.head()
```

## Mapping
Here I create a generic mapping function that will create a world map of something that varies by country, then use the function to visualize the cardiovascular disease data.

```python
#Mapping function 'map_quantity'
#Inputs
#country_vector:  List of country names (strings)
#quantity_vector:  Corresponding numerical array of quantities
#title:  Title for colorbar
def map_quantity(country_vector, quantity_vector, title):
    #Lists of uncommon country names and their replacements
    map_names_need_changed = [
    'Bolivia',
    "Cï¿½te d'Ivoire",
    'Republic of Congo',
    'United Kingdom',
    'The Gambia',
    'Iran',
    'Lao PDR',
    'Macedonia',
    'Dem. Rep. Korea',
    'Syria',
    'Tanzania',
    'United States',
    'Venezuela',
    'Vietnam']
    countries_to_replace = [
    'Bolivia (Plurinational State of)',
    "Cote d'Ivoire",
    'Democratic Republic of the Congo',
    'United Kingdom of Great Britain and Northern Ireland',
    'Gambia',
    'Iran (Islamic Republic of)',
    "Lao People's Democratic Republic",
    'The former Yugoslav republic of Macedonia',
    "Democratic People's Republic of Korea",
    'Syrian Arab Republic',
    'United Republic of Tanzania',
    'United States of America',
    'Venezuela (Bolivarian Republic of)',
    'Viet Nam']

    #Basic stats on data
    min_quant = min(quantity_vector)
    max_quant = max(quantity_vector)
    #Set up map elements and axes
    shapename = 'admin_0_countries'
    countries_shp = shpreader.natural_earth(resolution='110m',
                                            category='cultural', name=shapename)
    fig = plt.figure(figsize=(20,10))
    ax1 = fig.add_axes([0.05, 0.15, 0.9, 0.8], projection=ccrs.Robinson())
    ax2 = fig.add_axes([0.05, 0.05, 0.9, 0.07])
    #Loop through countries in map database
    for country in shpreader.Reader(countries_shp).records():
        this_country_name = country.attributes['name_long']
        #Change name if necessary
        if this_country_name in map_names_need_changed:
                this_country_name = countries_to_replace[map_names_need_changed.index(this_country_name)]
        #If map country is in list of quantities, assign color based on quantity and map it
        if this_country_name in country_vector:
            #Get quantity for this country
            this_quantity = quantity_vector[country_vector.index(this_country_name)]
            #Scale quantity on [0, 1]
            scaled_quant = (this_quantity - min_quant)/(max_quant - min_quant)
            #Get scaled color for map
            this_color = plt.cm.jet(int(round(scaled_quant*256.)))
            #Plot country with this color
            ax1.add_geometries(country.geometry, ccrs.PlateCarree(),
                              facecolor=this_color)
        else:
            #Plot country in gray
            ax1.add_geometries(country.geometry, ccrs.PlateCarree(),
                              facecolor=np.array([(0.5, 0.5, 0.5)]))
            #And print the country to know missing countries
            #print this_country_name

    #Create plot
    cmap = mpl.cm.jet
    norm = mpl.colors.Normalize(vmin=min_quant, vmax=max_quant)
    cb = mpl.colorbar.ColorbarBase(ax2, cmap=cmap,
                                    norm=norm,
                                    orientation='horizontal')
    font_size = 18
    cb.set_label(title,size=18)
    cb.ax.tick_params(labelsize=font_size)
    plt.show()
    return None
```

```python
map_quantity(cardio_2012['Country'].values.tolist(),
             cardio_2012['Age-standardized mortality rate by cause (per 100 000 population)'].values.astype('float'),
            'Age-standardized mortality rate of cardiovascular disease (per 100,000 population)')
```

We see that Eastern Europe and Russia seem to have the highest incidences, the Middle East has moderately high, and Western Europe, the Americas, Australia, and most Northern Asian countries relatively low rates of cardiovascular disease.

# Crop data
Data on food consumption is available from the United Nations Food and Agriculture Organization (http://www.fao.org/faostat/en/#home) non-animal products, here called crops, are available separately from animal products.  I load the data and perform some cleaning, and create a subset of data from the 10 years prior to 2000, the year in which I have disease data for model training.  This represents an assumption that a country's disease mortality rate in a given year is due to the national diet over the last decade.  Finally, I list all crop items in the data set.

# Combine food and disease data
Here I use compute simple statistics to represent average decadal food consumption, and peform an SQL style database merge to create the combined data set of food and disease.

```python
#Use pandas mean, which ignores NaNs, to get average decadal food consumption
meat_sums_2000 = meat_subset_2000.groupby(['AreaName', 'ItemName'], as_index=False).mean()
crop_sums_2000 = crop_subset_2000.groupby(['AreaName', 'ItemName'], as_index=False).mean()

#Get rid of meaningless sums
meat_sums_2000 = meat_sums_2000.drop(['AreaCode', 'ElementCode', 'ItemCode', 'Year'], axis=1)
crop_sums_2000 = crop_sums_2000.drop(['AreaCode', 'ElementCode', 'ItemCode', 'Year'], axis=1)

#"pivot" so that countries are rows and Item names are columns
meat_sums_2000 = meat_sums_2000.pivot(index='AreaName', columns='ItemName', values='Value')
crop_sums_2000 = crop_sums_2000.pivot(index='AreaName', columns='ItemName', values='Value')

#Repeat for testing data
meat_sums_2012 = meat_subset_2012.groupby(['AreaName', 'ItemName'], as_index=False).mean()
crop_sums_2012 = crop_subset_2012.groupby(['AreaName', 'ItemName'], as_index=False).mean()
meat_sums_2012 = meat_sums_2012.drop(['AreaCode', 'ElementCode', 'ItemCode', 'Year'], axis=1)
crop_sums_2012 = crop_sums_2012.drop(['AreaCode', 'ElementCode', 'ItemCode', 'Year'], axis=1)
meat_sums_2012 = meat_sums_2012.pivot(index='AreaName', columns='ItemName', values='Value')
crop_sums_2012 = crop_sums_2012.pivot(index='AreaName', columns='ItemName', values='Value')
```

```python
#Merge meat and crop into one food data set
all_food_2000 = pd.merge(meat_sums_2000, crop_sums_2000, right_index=True, left_index=True)
all_food_2012 = pd.merge(meat_sums_2012, crop_sums_2012, right_index=True, left_index=True)
print all_food_2000.shape, all_food_2012.shape
# all_food.head()
```

```python
# Set disease dataframe index as country
cardio_2000 = cardio_2000.set_index(['Country'])
cardio_2012 = cardio_2012.set_index(['Country'])
# Merge into final data frame
# Currently this throws out countries where the string name of the country doesn't match exactly
# Need to determine where this could be improved
result_2000 = pd.merge(all_food_2000, cardio_2000, right_index=True, left_index=True)
result_2012 = pd.merge(all_food_2012, cardio_2012, right_index=True, left_index=True)

#Change NaN to zero, representing assumption that unreported food items are of negligible consumption
result_2000.fillna(value=0, inplace=True)
result_2012.fillna(value=0, inplace=True)

#Put in table for regression
data_table_2000 = result_2000.values
data_table_2012 = result_2012.values
# np.shape(data_table_2000)
X_train = data_table_2000[:,0:-1]
Y_train = data_table_2000[:,-1]
X_test = data_table_2012[:,0:-1]
Y_test = data_table_2012[:,-1]

#Check item count
# result_2000.count()
```

```python
#Put all item names in an array
all_food_items_array = np.array(result_2000.columns.values)
# del all_food_items_list[-1]
all_food_items_array = np.delete(all_food_items_array, -1)
# print all_food_items_array
```

# Analysis
I try multiple linear regression as a simple initial approach, and lasso regression when that doesn't work out so well.

```python
#Based on http://scikit-learn.org/stable/modules/linear_model.html#ordinary-least-squares
from sklearn import linear_model
import sklearn
```

```python
clf = linear_model.LinearRegression()
clf.fit(X_train, Y_train)
print 'Linear regression'
print 'Train R^2 = ', clf.score(X_train, Y_train), ', Test R^2 = ', clf.score(X_test, Y_test)
```

While a linear regression using all 123 food items has a high training score, explaining 92% of the variance in the data, the testing score is negative, indicating the model describes the data no better than the average of all the responses would.  This indicates that the linear regression is overfit to the training data, and is not useful to predict the new, unseen observations in the testing data.  In statistical terms, it has high variance, and low bias.  Next I examine a lasso regression, which allows increased bias in the hopes that a lower variance will lead to less overfitting of the training data.

```python
clf2 = sklearn.linear_model.Lasso(max_iter=10000, alpha=1)
clf2.fit(X_train, Y_train)
print 'Lasso regression'
print 'Train R^2 = ', clf2.score(X_train, Y_train), ', Test R^2 = ', clf2.score(X_test, Y_test)
```

The lasso regression has a slightly lower training score, but a positive testing score, indicating it can explain nearly half the variance in the new, unseen data of the testing set.  This seems like a fairly successful performance, given the other factors such as exercise habits that may also affect rates of cardiovascular disease.

The lasso regression accomplishes reduced overfitting by penalizing the number of predictors and the magnitude of their coefficients.  I examine which coefficients were retained by the lasso, and their associations with cardiovascular disease.

```python
#Get coefficient values
coefs2 = clf2.coef_
#Table of coefficients
# zip(all_food_items_array, coefs2)
```

```python
#Make bar chart of non zero coefficients
non_zero_coeffs2 = coefs2[np.nonzero(coefs2)]
non_zero_items = all_food_items_array[np.nonzero(coefs2)]
# print np.shape(non_zero_coeffs2)
# print np.shape(non_zero_items)
fig = plt.figure(figsize=(20,10))
ax = plt.axes()

ind = np.arange(len(non_zero_coeffs2))  # the x locations for the groups
width = 0.5       # the width of the bars
ax.bar(ind, non_zero_coeffs2, width)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(non_zero_items)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=90, fontsize=12)
ax.grid(True)
plt.show
```

It appears that different types of oils have the largest positive and negative associations with cardiovascular disease.  In particular, palmkernel oil has the largest positive coefficient, implying that consumption of this increases the chance of disease, while fish body oil has the largest negative coefficient.
